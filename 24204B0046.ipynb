2. Sum of digits of a number

```python
num = int(input("Enter a number: "))
sum = 0

while num>0:
    sum+=num %10
    num //=10

print("Sum of digits =", sum)
```

3. Reverse a number

```python
num = int(input("Enter a number: "))
rev = 0

while num>0:
    rev = rev * 10 + num % 10
    num//=10
    print(rev)
```

4. Program to check if a number is a palindrome using for loop

```python
num = input("Enter a number: ")
rev = ""

for i in num:
    rev = i + rev

if num == rev:
    print("Palidrome number")
else:
    print("Not a Palidrome")
```

DO-WHILE SIMULATION

1. Keep taking input until the user enters 0

```python
while True:
    n = int(input("Enter a Number (0 to stop):"))
    if n==0:
        break
    print("you entered: ",n)
```

2. Menu-driven calculator

```python
while True:
    print("1.Add  2.Substract  3.Multiply  4.Exit")
    ch = int(input("Enter choice:"))

    if ch == 4:
        break

    a = int(input("Enter first Number: "))
    b = int(input("Enter second Number: "))
    
    if ch == 1:
        print("Sum = ",a+b)
    elif ch == 2:
        print("Difference = ",a-b)
    elif ch == 3:
        print("Product = ",a*b)
```

3. Continue asking for password until correct

```python
correct_password = "python"

while True:
    pwd = input("Enter Password: ")
    if pwd == correct_password:
        print("Access granted")
        break
    else:
        print("Incorrect password")
```
