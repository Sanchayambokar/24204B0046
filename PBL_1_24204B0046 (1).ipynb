 Evaluate arithmetic expressions and perform basic input/output operations

```python

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

addition = a + b
subtraction = a - b
multiplication = a * b
division = a / b

print("\nResults:")
print("Addition =", addition)
print("Subtraction =", subtraction)
print("Multiplication =", multiplication)
print("Division =", division)
```

 Calculate equivalent resistance of resistors connected in series and parallel

```python
r1 = float(input("Enter first resistance: "))
r2 = float(input("Enter second resistance: "))
req_series = r1 + r2
print("Equivalent Resistance in Series =", req_series, "ohms")

req_parallel = (r1 * r2) / (r1 + r2)
print("Equivalent Resistance in Parallel =", req_parallel, "ohms")

```

Calculate voltage using Ohm’s Law

```python
current = float(input("Enter current (in amperes): "))
resistance = float(input("Enter resistance (in ohms): "))

voltage = current * resistance

print("\nCalculated Voltage:")
print("Voltage =", voltage, "volts")
```

Check whether the entered frequency is audio or radio frequency

```python

frequency = float(input("Enter frequency in Hz: "))


if 20 <= frequency <= 20000:
    print("The entered frequency is an AUDIO frequency.")
elif frequency > 20000:
    print("The entered frequency is a RADIO frequency.")
else:
    print("The entered frequency is below the audible range (Infrasonic).")
```

 Display various radio frequency bands using if–elif ladder

```python
freq = float(input("Enter frequency in MHz: "))

if freq >= 3 and freq < 30:
    print("HF (High Frequency) Band")
elif freq >= 30 and freq < 300:
    print("VHF (Very High Frequency) Band")
elif freq >= 300 and freq < 3000:
    print("UHF (Ultra High Frequency) Band")
elif freq >= 3000 and freq < 30000:
    print("SHF (Super High Frequency) Band")
else:
    print("Frequency out of range")

```

Display resistor color code using match-case

```python
color = input("Enter resistor color: ").lower()

match color:
    case "black":
        print("Black → 0")
    case "brown":
        print("Brown → 1")
    case "red":
        print("Red → 2")
    case "orange":
        print("Orange → 3")
    case "yellow":
        print("Yellow → 4")
    case "green":
        print("Green → 5")
    case "blue":
        print("Blue → 6")
    case "violet":
        print("Violet → 7")
    case "grey" | "gray":
        print("Grey → 8")
    case "white":
        print("White → 9")
    case _:
        print("Invalid color entered")
```

Demonstrate looping constructs

```python
print("Loop Example:")
for i in range(1, 6):
    print(i)
```

```python
print("\nWhile Loop:")
i = 1
while i <= 5:
    print(i)
    i += 1
```

```python
print("\nNested Loop :")
for i in range(1, 4):
    for j in range(1, 4):
        print(i, j)
```

```python
print("\nBreak and Continue:")
for i in range(1, 6):
    if i == 3:
        continue
    if i == 5:
        break
    print(i)
```
