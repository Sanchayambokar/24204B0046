Write a Python program that checks if a resistor value is within a tolerance range (e.g., 5% tolerance for a 100Ω resistor) and prints whether it is within tolerance or not.

```python
nominal_resistance = 100
tolerance = 5

measured_resistance = 104

tolerance = nominal_resistance * (tolerance / 100)
lower_limit = nominal_resistance - tolerance
upper_limit = nominal_resistance + tolerance

if lower_limit <= measured_resistance <= upper_limit:
    print("The resistance is within the tolerance.")
else:
    print("The resistance is out within the tolerance.")
```
